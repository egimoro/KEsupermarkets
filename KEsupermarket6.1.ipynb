```python

# Apply a function with condition

def benchmark(x):
    if x > 1500:
        return x*.32
    else:
        return x*.68


np.round(df['change'].iloc[9:63].apply(benchmark), 2)
```

```python

np.round(df['total'].iloc[56:163].apply(benchmark), 2)
```
